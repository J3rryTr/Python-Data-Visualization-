```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_fwf('Homework_sampledata1.txt') # read data
df
```

```python
df = df.drop(['2008'], axis=1)
df
```

```python
with open('Homework_sampledata1.txt') as f:
    data = f.readlines()
for day, line in enumerate(data):
    if day == 1: 
        data = line.split() 
        line = line.strip()
    else: continue
    for month, value in enumerate(data):
        if month == 0: continue
        else: print(f'2008-{month}-{day} 00:00:00 {value}')
    break
```

```python
data1 = df.to_numpy() # convert to numpy array
data1
```

```python
data1[np.where(data1 == -99)] = np.nan
data1 
```

```python
data1 = data1.T.flatten()
print(data1)
```

```python
#plot data
plt.figure(figsize=(15, 5))
plt.plot(data1, "-b")
plt.xticks([0, 31, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335], ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"])
plt.ylabel("Temperature")
plt.title("Temperature in 2008")
plt.grid(True)
plt.xlim(0, 365)
plt.ylim(5, 35)
plt.legend(["Temperature"])
plt.show()
```
