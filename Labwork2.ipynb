```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_fwf('Homework_sampledata1.txt')
df
```

```python
df_drop = df.drop(['2008'], axis=1) # drop column 2008
month =['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
df_drop.columns = month
df_drop
```

```python
data = df_drop.to_numpy() # convert pandas table to numpy array
data
```

```python
data[data == -99] = np.nan # replace -99 by NaN
print(data)
```

```python
data1 = data.T.flatten()
data1 = data1[~np.isnan(data1)]

print(len(data1))
print(data1)
```

1. Histogram

```python
# Histogram
plt.figure(figsize=(15, 8)) # plot
plt.hist(data1, 50, density=True)
plt.title('Histogram', size=20)
plt.ylabel('Probability density', size=15)

#KDE
from sklearn.neighbors import KernelDensity
kde = KernelDensity(kernel='gaussian', bandwidth=1).fit(data1.reshape(len(data1),1))  #This function only accepts 2D array
x_axis = np.linspace(np.min(data1), np.max(data1), 200)
log_dens = kde.score_samples(x_axis.reshape(200,1))      #This function only accepts 2D array

plt.plot(x_axis, np.exp(log_dens), color='r', lw=1, linestyle='-', label='gaussian kernel',)
plt.legend()
plt.show()
```

2. Boxplot

```python
# Boxplot
fig,axes = plt.subplots(figsize=(10,8))
fig= sns.boxplot(df_drop, ax=axes)
axes.set_xticklabels(labels= df_drop.columns)
axes.set_title('Boxplot')
axes.set_xlabel('Month')
axes.set_ylabel('Temperature (oC)')
```

3. Violin plot

```python
# violin plot

# convert numpy array to list
df_list = list(data.T)
new_df_list = []
for i in range(len(df_list)):
    without_nan = [x for x in df_list[i] if str(x) != 'nan']
    new_df_list.append(without_nan)

plt.figure(figsize=(15, 8))
plt.violinplot(new_df_list, showmeans= False, showmedians= False, showextrema= False)
for i in range(len(new_df_list)):
    plt.vlines(i+1, np.min(new_df_list[i]), np.max(new_df_list[i]), color='k', linestyle='-', lw=1)
    plt.vlines(i+1, np.percentile(new_df_list[i], 25), np.percentile(new_df_list[i], 75), color='k', linestyle='-', lw=5)
    plt.scatter(i+1, np.median(new_df_list[i]), marker='o', color='white', s=20, zorder=3)
plt.title('Violin plot')
plt.xlabel('Month')
plt.xticks(range(1, 13), df_drop.columns)
plt.ylabel('Temperature (oC)')
plt.show()
```

4. Error bar

```python
# calculate mean and standard deviation
mean = df_drop.mean()
std = df_drop.std()
```

```python
# plot
subtract = list(map(lambda x, y: x - y, mean, std))
add = list(map(lambda x, y: x + y, mean, std))
plt.figure(figsize=(15, 8))
plt.bar(df_drop.columns, mean, yerr=std, color='aqua')
plt.xticks(df_drop.columns)
plt.vlines(df_drop.columns, subtract, add, color='k', linestyle='-', lw=1)
plt.ylim(10, 30)
plt.title('Error plot', size=20)
plt.xlabel('Months', size=15)
plt.show()
```
